```python
#Importando a blibioteca pandas e importando a base de dados do Excel
import pandas as pd



```

```python
#Importando a base de dados do Excel e mostrando as 10 primeiras linhas 
combustiveis_df = pd.read_excel('ca-2021-02.xlsx')
display(combustiveis_df.head(10))
```

```python
#Inserção de dados simples ( Criando uma nova coluna)

combustiveis_df['Ativo'] = True
display(combustiveis_df.head(10))
```

```python
#Cria uma coluna "Obs" que tenha escrito 'MELHOR CIDADE ' quando a coluna municipio for igual a 'INDAIATUBA'

combustiveis_df['Obs'] = ["MELHOR CIDADE" if municipio == 'SAO PAULO' else None for municipio in combustiveis_df['Municipio']]

display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO','INDAIATUBA', 'CAMPINAS', 'SALTO']), ['Municipio', 'Obs']])
```

```python
#Usando a blibioteca Numpy
#Acrescentando uma coluna 'STATUS' e colocando 'Caro' para valores de gasolina acima de 6,50 e abaixo colocando "Caro"
#Copiei do Leandro Rodrigues rsrs


import numpy as np

combustiveis_df['Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display(combustiveis_df[['Valor de Venda', 'Status']])
 
```

```python
#Importando a abse de dados e separanaod os dados por coluna com base no ';'
num_habitantes_df= pd.read_csv('ibge_num_habitantes_estimado.csv', sep = ';')
display(num_habitantes_df)
```

```python
#Renomear a coluna para as duas bases de dados terem o mesmo nome

num_habitantes_df = pd.read_csv("ibge_num_habitantes_estimado.csv", sep=";")
num_habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
#Cruzar as duas bases de dados (MERGE)
#Coloca os dois data frames ,how = varias possibilidades ( a que usarei sera 'left' ), on = são as colunas
#O left considera o data frame da esquerda como forte e ao comparar os dados , se nao estiver igual , ele priorizara os dados da esquerda e deixara 'Nulo' as linhas em branco

#colunas = ['Municipio','Estado - Sigla']
#merge_df = combustiveis_df.merge(num_habitantes_df, how = 'left' , on= colunas)
#display(merge_df)
#print(merge_df.info())
```

```python
#Cruzar as duas bases de dados (MERGE)
#Coloca os dois data frames ,how = varias possibilidades ( a que usarei sera 'outer' ), on = são as colunas
#Ele mistura tudo (left, rigth e inner'), se tiver a cidade , ele priorizara tanto na esquerda quanto na direita , e se nao tiver nada , ele ainda mostrara as cidades porem sem dados

#colunas = ['Municipio','Estado - Sigla']
#merge_df = combustiveis_df.merge(num_habitantes_df, how = 'outer' , on= colunas)
#display(merge_df)
#print(merge_df.info())
```

```python
#Cruzar as duas bases de dados (MERGE)
#Coloca os dois data frames ,how = varias possibilidades ( a que usarei sera 'inner' ), on = são as colunas
#O inner deixou apenas os dados que tem os municípios do df num habitantes

colunas = ['Municipio','Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes_df, how = 'inner' , on= colunas)
display(merge_df)
print(merge_df.info())
```

```python
#Iremos 'destruir' as colunas que não iremos ultilizar
#dropna (ira destruir a colunas ou linhas que estiverem totalmete vazias (nulas) 
# 'axis = 1' ou axis = 'columns'  ira para excluir as colunas e 'axis = 0 ' ou  'axis = index 'ira excluir as linhas

merge_df.dropna( axis = 1 , inplace = True)
display(merge_df.info())
```

```python
#Eliminando todas as colinas que nao iremos ultilziar
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo', 'Status']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
#Eliminas as linhas duplicadas

merge_df.drop_duplicates(inplace=True)
display(merge_df)
```

```python
#Agrupar e contar quantos postos tem na cidade

postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
display(postos_por_municipio_df.info())
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "Número de Postos"}, inplace=True)
display(postos_por_municipio_df)
```

```python
#Calcular postos de gasolina X habitantes temos na amostragem de combustiveis nov/2021
print(postos_por_municipio_df.info())
postos_por_municipio_df['Postos por Habitante'] = postos_por_municipio_df['Numero de Postos']/ postos_por_municipio_df['NumHabitantes2021']
display(postos_por_municipio_df)

#Final com problema
```
